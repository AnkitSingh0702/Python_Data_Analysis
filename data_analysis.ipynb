```python
# import all the required library
import numpy as np 
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# loading the dataset 
df =  pd.read_csv('Diwali_Sales_Data.csv')
df
```

```python
# It will tell the total number of rows and columns in the dataset 
df.shape
```

- **Now we will clean the data** 

```python
# checking the information of the datset 
# here in the dataset we can see that status and the unnamed1 as 0 value in the non null count
df.info()
```

```python
# so as we see in the info status and unnamed containes no value so we will drop that values from the dataset
# axis=1 means we are dropping columns (not rows)
# inplace=True means the changes are made directly to the original dataframe
df.drop(['Status' , 'unnamed1'] , axis = 1  , inplace= True )
```

```python
df.info()
```

```python
# Now we have to check the null values in the dataset 
df.isnull().sum()
# as we can see there is null values in Amount section 
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
# if i want to covert the datatype in a dataset than we have to 
# if we want to change the datatype we use .astype('datatype in which you want to convert') 
df['Amount']= df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
# .columns it will show all the columns present in the dataset 
df.columns
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
```

```python
# if we want to see the describe feature of specific columns than we want to do 
df[['Age']].describe()
```

# **Exploratory Data Analysis**

```python
df.columns
```

### **Analysing the Gender** 

```python
ax  =  sns.countplot( x = 'Gender' , data= df)


# ax.containers is a list of BarContainer objects, each representing the bars in the countplot.
# We iterate over these containers to add labels to each bar.
for i in ax.containers:
    ax.bar_label(i)
    
# from this graph we can conclude that the number of female orders more as compare to male 
```

-  **plotting a bar chart for gender vs total amount**

```python
# First, we are grouping the 'Amount' column on the basis of 'Gender' using groupby.
# The groupby(['Gender'], as_index=False) groups the data by the 'Gender' column and keeps 'Gender' as a column (not index).
# ['Amount'].sum() calculates the total purchase amount for each gender.
sales_gen = df.groupby(['Gender'] , as_index= False)['Amount'].sum()
sales_gen
```

```python
sns.barplot( x = 'Gender' , y = 'Amount' , data = sales_gen)
```

- From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

### **AGE**

```python
df.columns
```

```python
# hue : The 'hue' parameter in seaborn plots is used to add a third categorical variable by grouping data within the main categories and displaying them in different colors.
# In this case, using hue='Gender' allows us to compare the counts of each gender within every age group, making it easy to visualize gender distribution across age groups.
ax = sns.countplot(data = df , x = 'Age Group' , hue = 'Gender')

for i in ax.containers:
    ax.bar_label(i)
```

- **Total amount vs age group**

```python
sales_age = df.groupby(['Age Group'] , as_index= False)['Amount'].sum().sort_values(by = 'Amount' , ascending= False)
sales_age
```

```python
sns.barplot(data = sales_age , x = 'Age Group' , y ='Amount')

# From the graph we can conclude that the most of the buyers are of age group between 26-35 years female
```

### **State**

- **Total number of orders from top 10 states**

```python
# Grouping the data by 'State', summing the 'Orders' for each state, sorting in descending order, and selecting the top 10 states
sales_state = df.groupby(['State'] , as_index= False )['Orders'].sum().sort_values(by = 'Orders' , ascending= False).head(10)
sales_state
```

```python
plt.rcParams['figure.figsize'] = [25,5]
sns.barplot(data = sales_state , x = 'State' , y = 'Orders')

# From this graph we can observe that Uttar Pradesh has the most active buyers in this dataset.
```

- **Total amount/sales from top 10 states**


```python
sales_amount_state = df.groupby(['State'] , as_index= False )['Amount'].sum().sort_values(by = 'Amount' , ascending= False).head(10)
sales_amount_state
```

```python
plt.rcParams['figure.figsize'] = [25,5]
sns.barplot(data = sales_amount_state , x = 'State' , y = 'Amount')
```

- From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

### **Marital Status**

- ***This plot is on the basis of count of Marital_Status***

```python
# This plot is on the basis of count of Marital_Status
plt.rcParams['figure.figsize'] = [7,5]
ax = sns.countplot(data = df , x = 'Marital_Status')

for i in ax.containers:
    ax.bar_label(i)

# so from the graph we can conclude that marid shops more 
```

- ***Analyze the total purchase amount based on Marital Status and Gender using a bar plot.***

```python
# Analyze the total purchase amount based on Marital Status and Gender using a bar plot.
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.rcParams['figure.figsize'] = [6,5]
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

- From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

### Ocuupation

```python
df.columns
```

```python
plt.rcParams["figure.figsize"] = [20 ,5]

ax  = sns.countplot(data = df , x = 'Occupation')

for i in ax.containers:
    ax.bar_label(i)
```

```python
# Analyze the impact of different occupations on the total purchase amount during Diwali sales.
# Group the dataset by 'Occupation', calculate the total 'Amount' spent by each occupation,
# and visualize the results using a bar plot to identify which occupations contribute most to the sales.
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.rcParams["figure.figsize"] = [20 , 5]
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

- From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

# Product category

```python
df.columns
```

```python
# Distribution of Transactions Across Product Categories

plt.rcParams["figure.figsize"] = [26 , 5]
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Top 10 Product Categories by Total Sales Amount
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.rcParams["figure.figsize"] = [20 , 5]
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

-  From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
# Top 10 Most Ordered Products
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

plt.rcParams["figure.figsize"] = [20 , 5]
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
# top 10 most sold products
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
